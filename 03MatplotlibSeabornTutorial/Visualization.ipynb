```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x, y)
# plt.show()
```

```python
plt.plot(x, y)
plt.xlim(0, 2)
plt.ylim(100, 300)

plt.title("Straight Line")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
plt.plot(x, y, color='red')
plt.xlim(0, 2)
plt.ylim(100, 300)

plt.title("Straight Line")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()

plt.savefig('quantityprice.png')
```

```python
plt.plot(x, y, color='red', marker='o', markersize=20, linestyle="--")
plt.xlim(0, 2)
plt.ylim(100, 300)

plt.title("Straight Line")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# create a canvas to plot the graph using the method subplots()
# you can even create more than one graph in one canvas using this subplots method
# which returns 2 values figure and axis
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='About as simple as it gets, folks')
ax.grid()

# you can save the figure as an image using savefig method
fig.savefig("test.png")
plt.show()
```
