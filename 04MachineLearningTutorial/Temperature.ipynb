```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
celsius = np.arange(0, 101, 10)
celsius
```

```python
fahrenheit = (celsius * 9/5) + 32
fahrenheit
```

```python
plt.scatter(celsius, fahrenheit)
```

```python
df = pd.DataFrame(celsius, columns=['celsius'])
df['fahrenheit'] = fahrenheit
df
```

```python
sns.regplot(x = 'celsius', y='fahrenheit', data = df)
```

```python
# y = mx + c
# x => celsius value
# y => farenheit value
df = sns.load_dataset('iris')
df
```

```python
sns.regplot(x = 'sepal_length', y='petal_length', data = df)
```

```python
# Linear regression Model
# X => Independent Variables (celsius) (sepel_length, petal_length)
# y => Dependent variable (farenheit) (species => setosa)
X = celsius.reshape(-1, 1) # row, column (-1 means None)
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create instance of the model
model = LinearRegression()
```

```python
# Let us train it
model.fit(X, fahrenheit)
# model.fit(df[['A', 'B', 'C']], df['D'])
```

```python
celsius_new = np.array([25, 60, 95])
X_new = celsius_new.reshape(-1, 1)
X_new
```

```python
model.predict(X_new)
```

```python
# slop => m
model.coef_[0]
```

```python
# intercept => c
model.intercept_
```
