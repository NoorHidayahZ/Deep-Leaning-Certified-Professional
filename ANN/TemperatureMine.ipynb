```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
```

```python
celsius = np.arange(0, 101, 10  )
celsius
```

```python
fahrenheit = (celsius *9/5) + 32
fahrenheit
```

```python
X_train = celsius
y_train = fahrenheit
```

```python
#instance
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))  #units = 1, input_shape = 1
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
```

```python
model.summary()
```

```python
epochs_hist = model.fit(X_train, y_train, epochs=500)
```

```python
epochs_hist.history
```

```python

model.predict([60, 77, 110])
```
